# Exploratory Data Analysis

# Age

```python
# plotting a bar chart for Gender and it's count
```

```python
sns.countplot(x="Gender",data=df)
```

```python
sns2=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount
```

```python
sns.barplot(x="Gender",y="Amount",data=sns2)
```

From the above bar we can see that most of the purscahing has been done my female 

```python
sns.countplot(x="Age Group",hue="Gender",data=df)
```

```python
# Total Amount vs Age Group
```

```python
sns3=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns3
```

```python
sns.barplot(x="Age Group",y="Amount",data=sns3)
```

From the  ghraps we can undertsnad that most of the buyers are from the age group of 26-35 years 

# State

```python
# total number of orders from top 10 states
```

```python
df["State"].unique()
```

```python
sns4=df.groupby(["State"],as_index=False).Orders.sum().sort_values(by="Orders",ascending=False).head(10)
sns4
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Orders",data=sns4)

```

```python
sns5=df.groupby(["State"],as_index=False).Amount.sum().sort_values(by="Amount",ascending=False).head(10)
```

```python
sns.barplot(x="State",y="Amount",data=sns5)
```

```python
df.head(5)
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
